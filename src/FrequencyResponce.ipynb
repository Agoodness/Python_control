# Explicit frequency responce 

```python
import control as ctrl
import matplotlib.pyplot as plt
import numpy as np
```

```python
s = ctrl.TransferFunction.s
G = 3 / (s**2 + s*(1/2) + 4)

print("Transfer function G(s):")
print(G)
```

```python
omega = np.logspace(-2, 2, 500)
response = ctrl.freqresp(G, omega)
Gjw = response.response 

real_parts = np.real(Gjw)
imag_parts = np.imag(Gjw)

# Convert to magnitude and phase
magnitude = np.abs(Gjw)
magnitude = 20 * np.log10(magnitude)  # Convert to dB
phase = np.angle(Gjw, deg=True)

# Plot Bode plots
plt.figure()
plt.semilogx(omega, magnitude)
plt.ylabel("Magnitude (dB)")
plt.grid(True)

plt.figure()
plt.semilogx(omega, phase)
plt.ylabel("Phase (deg)")
plt.xlabel(" (rad/s)")
plt.grid(True)
plt.show()

print("Frequency response at specific frequencies:")
print("Frequencies (rad/s):", real_parts)


```

```python
# mag, phase, omega = ctrl.bode(G, omega, dB=True, Plot=True
mag, phase, Omega = ctrl.bode_plot(G, omega, dB=True, plot=False)

mag_db = 20 * np.log10(mag)
phase_deg = np.degrees(phase)
plt.figure()
plt.semilogx(omega, mag_db)
plt.ylabel("Magnitude (dB)")
plt.grid(True)

plt.figure()
plt.semilogx(omega, phase_deg)
plt.ylabel("Phase (deg)")
plt.xlabel(" (rad/s)")
plt.grid(True)
plt.show()
```
